```python
def f(x):
    return 1/6*x**6-3/5*x**5-x**4+4*x**3
```

```python
def df(x):
    return x**5-3*x**4-4*x**3+12*x**2
```

```python
eps=0.01;alpha=0.1;Iter=1000;
```

```python
x = 3
for i in range(Iter+1):
    x = x-alpha*df(x)
    print("i={}, x={}, f(x)={}".format(i,x,f(x)))
    if abs(df(x)) < eps:
        break
```
