```python
from urllib.parse import urljoin
import pandas as pd
base_url = "https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/"
anime_csv = urljoin(base_url, "anime.csv")
df = pd.read_csv(anime_csv)
df.head()
```

```python
df = pd.read_csv(anime_csv, index_col=0)
df.head()
```

```python
df = pd.read_csv(anime_csv, dtype={"members": float})
df.dtypes
```

```python
anime_stock_price_csv = urljoin(base_url, "anime_stock_price.csv")
df = pd.read_csv(anime_stock_price_csv, parse_dates=["Date"])
df.dtypes
```

```python
anime_tsv = urljoin(base_url, "anime.tsv")
df = pd.read_csv(anime_tsv, sep="\t")
df.head()
```

```python
# 列ごとの平均値を算出
from urllib.parse import urljoin
import pandas as pd
base_url = (
 "https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/"
)
anime_master_csv = urljoin(base_url, "anime_master.csv")
df = pd.read_csv(anime_master_csv)
# 列ごとの平均値を算出
df.mean(numeric_only=True)
```

```python
# Series の合計値を算出
df["members"].sum()
```

```python
df.describe().round(1)
```

```python
df.describe(percentiles=[0.1, 0.9]).round(1)
```
